```python
import pandas as pd

df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)

print(df.head())
```

```python
import matplotlib.pyplot as plt

# Top 10 countries by population in 2024
top10 = df[['Country Name', '2024']].dropna().sort_values(by='2024', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2024'])
plt.title('Top 10 Countries by Population (2024)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
